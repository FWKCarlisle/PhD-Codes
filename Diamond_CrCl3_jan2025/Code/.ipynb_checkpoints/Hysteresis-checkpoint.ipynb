```python
def make_file_list(numbers):
    files = []
    for number in numbers:
        files.append(f'i06-1-{number}.nxs')
    return files

print(make_file_list([1,2,3]))
```

```python
# Insert your folder path here. Ideally just place your data in a subfolder
# so you don't have to copy the whole path
folder = ""
numbers = [329930]
file_list = make_file_list(numbers)
print(file_list)
```
